# sample data load

```python
import os
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
path= 'C:/Users/baek/Desktop/NURO/sample'
file_list= os.listdir(path)
```

```python
dat = [0 for i in range(0,2)]

for i in range(0,2):
    dat[i] = pd.read_csv(path+"/"+file_list[i], sep = ',')
```

# data visualization

```python
for j in range(len(dat)):
        plt.figure(figsize=(30, 15))
        for i in range(1,3):
            plt.plot(dat[j].loc[dat[j]['CoreSN'] == 'core0'+str(i),'ActualTorque'],label='CORE'+str(i))
        title_font = {
            'fontsize' : 32,
            'fontweight' : 'bold'
        }
        plt.title('ActualTorque '+ str(j+1), fontdict=title_font)
        plt.ylim([-22, 22])
        plt.legend(fontsize=20)
        plt.show()
```
